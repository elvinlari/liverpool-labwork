### Image Analysis by Numpy Array
Uploaded image in Python and converted into a NumPy array

```python
# Import Libraries
%pip install Pillow
from PIL import Image
import numpy as np
%pip install matplotlib
import matplotlib.pyplot as plt
```

```python
# Load image
image_path = '/home/elvin/DEVELOPMENT/liverpool/newlabwork/test_driving_lc_sample.jpg'
image = Image.open(image_path)

# Convert the image to a NumPy array
image_array = np.array(image)

# Display the shape of the array (height, width, channels)
image_array_shape = image_array.shape

# Print the image array with improved formatting
print("Image Array (Sample of Pixel Values):")
print(image_array)

print("\nArray Shape (Height, Width, Channels):")
print(image_array_shape)
```

### Pixel Statistics

Analyze basic statistics like mean, minimum, and maximum pixel values for each channel or grayscale image.


```python
# Basic statistics
mean_pixel_value = np.mean(image_array_shape)
min_pixel_value = np.min(image_array_shape)
max_pixel_value = np.max(image_array_shape)

print("Image Statistics:","\nMean:",mean_pixel_value,"Min:",min_pixel_value,"Max:",max_pixel_value)
```


### Color Channel Analysis

For RGB images, analyze the individual red, green, and blue channels.


```python
# Extract individual channels
R, G, B = image_array[..., 0], image_array[..., 1], image_array[..., 2]

# Calculate the mean pixel intensity for each channel
mean_R = np.mean(R)
mean_G = np.mean(G)
mean_B = np.mean(B)

print("RGB Channel Mean Intensity:", "\nRed:", mean_R,"Green:", mean_G, "Blue:",mean_B)
```


### Convert colored image to grayscale image using the NumPy array

Apply a formula that calculates the weighted average of the red, green, and blue channels.

It is a standard approach for converting to grayscale because it considers human perception, giving different weights to each color channel (typically 0.2989 for red, 0.5870 for green, and 0.1140 for blue).

### What is uint8 in Images?

uint8 stands for "unsigned 8-bit integer", which is a data type commonly used in digital image processing to represent pixel intensity values. In this format:

Each pixel is stored as an integer between 0 and 255. The uint8 data type uses 8 bits (1 byte) per pixel, allowing for 256 unique values.


```python


# Extract the RGB channels
R, G, B = image_array[..., 0], image_array[..., 1], image_array[..., 2]

# Apply the grayscale formula
grayscale_image_array = (0.2989 * R + 0.5870 * G + 0.1140 * B).astype(np.uint8)

# Display the grayscale image
plt.imshow(grayscale_image_array, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')  # Turn off-axis labels for better visualization
plt.show()

```


### What is a Binary Image?

A binary image is a digital image where each pixel has only two possible intensity values, typically 0 (black) and 255 (white). Binary images are often used in computer vision and image processing tasks for object detection, segmentation, and simplification of visual data.

The process of creating a binary image is called thresholding, where a threshold value is selected, and each pixel in the grayscale image is compared against this threshold:

    Pixels with intensity greater than the threshold are set to 255 (white).
    Pixels with intensity less than or equal to the threshold are set to 0 (black).



```python
# Define a threshold for the binary conversion
threshold = 50  # You can adjust this value based on the image
binary_image_array = (grayscale_image_array > threshold).astype(np.uint8) * 255

# Display the binary image
plt.imshow(binary_image_array, cmap='gray')
plt.title("Binary Image")
plt.axis('off')  # Turn off axis labels for better visualization
plt.show()
```

```python
# Define a threshold for the binary conversion
threshold = 150  # You can adjust this value based on the image
binary_image_array = (grayscale_image_array > threshold).astype(np.uint8) * 255

# Display the binary image
plt.imshow(binary_image_array, cmap='gray')
plt.title("Binary Image")
plt.axis('off')  # Turn off axis labels for better visualization
plt.show()
```
